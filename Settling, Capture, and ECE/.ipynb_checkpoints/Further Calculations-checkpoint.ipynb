```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from matplotlib import interactive
interactive(True)
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import seaborn as sns
sns.set()
import statsmodels.api as sm
from scipy import interpolate
from scipy import stats
from scipy import integrate
from scipy.optimize import curve_fit
import matplotlib.cm as cm
import warnings
warnings.filterwarnings('ignore')
```

```python
def reynolds_number(freq): #frequency of flume, at either 10, 20, or 30 Hz
    velocity = 0.00196*freq + 0.00081
    return round((velocity*0.003175)/(9.509e-7), 3), velocity

def frontal_area(density):
    return round((density*1.2*0.003175*0.4)/(0.4*1.2), 3)

#test section settling based on averaged sediment trap values across entire test section
def test_section_settling(csv_name, plot = 0):
    csv_name = csv_name[0:6] + 'sedtrap.csv'
    sample = pd.read_csv(csv_name)
    sample = sample[0:9]
    indices = list(sample[sample['Sediment'].notnull()].index)
    sample = sample.iloc[indices,:]
    sediment_settled = (np.mean(sample['Sediment']*0.001)*1.95*0.6)/(np.pi*0.0127**2)
    return sediment_settled

run_variables = pd.read_csv('ece_table.csv')
```

```python
low_re_c = run_variables[(run_variables['Velocity'] == 10) & (run_variables['Density'] != 0)]
med_re_c = run_variables[(run_variables['Velocity'] == 20) & (run_variables['Density'] != 0)]
high_re_c = run_variables[(run_variables['Velocity'] == 30) & (run_variables['Biofouled'] == 0) & (run_variables['Density'] != 0)]
```

```python
plt.figure(8, figsize=(10,7))
sns.lineplot(low_re_c['Frontal Area'], low_re_c['ECE %'], marker = 'o', color = '#32a887');
sns.lineplot(med_re_c['Frontal Area'], med_re_c['ECE %'], marker = 'o', color = '#5bbefc');
sns.lineplot(high_re_c['Frontal Area'], high_re_c['ECE %'], marker = 'o', color = '#f2be63');
plt.xlabel('Frontal Area ($m^{2}/m^{3}$)');
plt.ylabel('Estimated Collector Efficiency (%)');
plt.legend(['Low: 64.48', 'Medium: 126.4', 'High: 188.32'], title = '$Re_{c}$');
```

```python
with_dowel_runs = run_variables[(run_variables['Density'] != 0) & (run_variables['Biofouled'] == 0)]
dimless_dowel = with_dowel_runs['Density']*0.003175**2
mult = with_dowel_runs['Re_c']*dimless_dowel
plt.figure(figsize=(10,7))
sns.lineplot(mult, with_dowel_runs['ECE %'], marker = 'o', color = '#f2be63');
plt.xlabel('$Re_{c} * Dimensionless\ Dowel\ Density$');
plt.ylabel('Effective Collector Efficiency (ECE) %');
plt.title('$Re_{c} * Dimensionless\ Dowel\ Density$ vs. ECE %');
```

```python
def power_function(x, a, b):
    return a * np.power(x, b)

x = mult
y = with_dowel_runs['ECE %']
popt_power, pcov_power = curve_fit(power_function, x, y)

#calculating r-squared
residuals = y - power_function(x, popt_power[0], popt_power[1])
ss_res = np.sum(residuals**2)
ss_tot = np.sum((y-np.mean(y))**2)
r_squared = 1 - (ss_res / ss_tot)
r_squared
```

```python
plt.figure(figsize= (10,7))
sns.scatterplot(x, residuals, s = 80);
plt.axhline(0, color='black', ls = '--');
plt.xlabel('$Re_{c} \cdot$ Dimensionless Dowel Density');
plt.ylabel('Residual');
plt.title(f"Residuals of Fitted Power Law, $r^{2} = {round(r_squared, 3)}$");
```

```python
#plot fit
xs = np.arange(min(mult), max(mult), (max(mult)- min(mult))/100)
ys = power_function(xs, popt_power[0], popt_power[1])
frontal_area_str = with_dowel_runs['Frontal Area'].astype(str)
fig = plt.figure(10, figsize=(10,7))
ax = sns.scatterplot(mult, with_dowel_runs['ECE %'], marker = 'o', s = 80, hue = frontal_area_str, palette = ["#ffc7e2", '#ff47a0', '#a30351'], style = with_dowel_runs['Re_c'])
sns.lineplot(xs, ys, color = '#f2be63');
plt.xlabel('$Re_{c} \cdot$ Dimensionless Dowel Density');
plt.ylabel('Effective Collector Efficiency (ECE) %');
plt.title('$Re_{c} \cdot$ Dimensionless Dowel Density vs. ECE %');
handles, labels = ax.get_legend_handles_labels()
area_legend = ax.legend(handles=handles[1:4], labels=labels[1:4], title="Frontal Area per \n Unit Vol. ($m^{-1}$)", loc = (0.812,0.76));
re_c_legend = ax.legend(handles=handles[5:8], labels=labels[5:8], title="$Re_{c}$", loc = (0.85, 0.55));
ax.add_artist(area_legend);
ax.add_artist(re_c_legend);
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
def cross_sectional_dowel_area(frontal_area):
    dowel_surface_area = frontal_area*0.4*0.6*0.003175
    return dowel_surface_area

def discharge_obstructed(velocity, area):
    return reynolds_number(velocity)[1]*area
```

```python
discharge_array = []
for i in range(len(with_dowel_runs)):
    cs_dowel_area = cross_sectional_dowel_area(with_dowel_runs['Frontal Area'].iloc[i])
    flow_obstructed = discharge_obstructed(with_dowel_runs['Velocity'].iloc[i], cs_dowel_area)
    discharge_array.append(flow_obstructed)
```

```python
#fit mass-flow data points to power law function
x = discharge_array
y = with_dowel_runs['ECE %']
popt_power_discharge, pcov_power_discharge = curve_fit(power_function, x, y)

#calculating r-squared
residuals = y - power_function(x, popt_power_discharge[0], popt_power_discharge[1])
ss_res = np.sum(residuals**2)
ss_tot = np.sum((y-np.mean(y))**2)
r_squared = 1 - (ss_res / ss_tot)
```

```python
xs = np.arange(min(discharge_array), max(discharge_array), (max(discharge_array)- min(discharge_array))/100)
ys = power_function(xs, popt_power_discharge[0], popt_power_discharge[1])
fig = plt.figure(10, figsize=(10,7))
ax = sns.scatterplot(discharge_array, with_dowel_runs['ECE %'], s=80, color = '#a30351');
sns.lineplot(xs, ys, color = '#f2be63');
plt.xlabel('Discharge Altered by Vegetation ($m^{3}/s$)');
plt.ylabel('Effective Collector Efficiency (ECE) %');
plt.title('Discharge Altered by Vegetation vs. ECE %');
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim([min(xs) - 0.002, max(xs) + 0.0001])
ax.text(x = 0.0002, y = 0.63, s = f"$r^{2}$ = {round(r_squared,3)}");
```

```python
def mass_collected(k_c, k_t):
    m_c = (k_c*200*(1-np.exp(-k_t*6000)))/k_t
    return m_c
```
