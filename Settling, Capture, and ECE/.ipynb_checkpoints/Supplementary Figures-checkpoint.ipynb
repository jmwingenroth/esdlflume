```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from matplotlib import interactive
interactive(True)
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import seaborn as sns
sns.set()
import statsmodels.api as sm
from scipy import interpolate
from scipy import stats
from scipy import integrate
import matplotlib.cm as cm
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.style.use('seaborn-whitegrid')
x_pos = np.array([0.15, 0.15, 0.15, 0.745, 0.93, 1.15, 1.75, 1.75, 1.75])
y_pos = np.array([0.11, 0.3, 0.51, 0.44, 0.3, 0.14, 0.11, 0.3, 0.51])
labels = ['I', 'H', 'G', 'F', 'E', 'D', 'C', 'B', 'A']
fig, ax = plt.subplots(figsize=(13.65,4.2))
sns.scatterplot(x_pos, y_pos, s = 120);
plt.xlim([0,1.95]);
plt.ylim([0,0.6]);
plt.xlabel('Distance from Downstream Edge of Test Section (m)', size = 13);
plt.ylabel('Distance from Left Wall of Flume (m)', size = 13);
style = dict(size=16, color='black', weight = 'bold')
for i in range(len(x_pos)):
    ax.text(x_pos[i], y_pos[i], labels[i], position = (x_pos[i]+0.02, y_pos[i]+0.01))
```
