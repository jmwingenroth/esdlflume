```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

import pandas as pd

from scipy.io import loadmat

from pyod.models.knn import KNN
from pyod.models.auto_encoder import AutoEncoder

from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
#loading data
file = loadmat('30Hz_OutsideDowels.mat')
high_vel_out = file['Data\x00\x00\x00\x00']
vel_X = high_vel_out['Profiles_VelX']
vel_Y = high_vel_out['Profiles_VelY']
vel_Z1 = high_vel_out['Profiles_VelZ1']
vel_Z2 = high_vel_out['Profiles_VelZ2']
time = high_vel_out['Profiles_TimeStamp']
```

```python
#universal maximums
vel_Y_avg = np.mean(vel_Y[0][0], axis = 1)
std_Y_avg = np.std(vel_Y_avg)
vel_Y_max = std_Y_avg*(2*np.log(len(vel_Y_avg)))**0.5

vel_Z1_avg = np.mean(vel_Z1[0][0], axis = 1)
std_Z1_avg = np.std(vel_Z1_avg)
vel_Z1_max = std_Z1_avg*(2*np.log(len(vel_Z1_avg)))**0.5

vel_Z2_avg = np.mean(vel_Z2[0][0], axis = 1)
std_Z2_avg = np.std(vel_Z2_avg)
vel_Z2_max = std_Z2_avg*(2*np.log(len(vel_Z2_avg)))**0.5
```

```python
pure = np.zeros(len(vel_Y_avg))
noise = np.random.normal(0, vel_Y_max**2, pure.shape)
signal = pure + noise
train = pd.DataFrame(signal).rename(columns = {0:'Y'})
test = pd.DataFrame(vel_Y_avg).rename(columns = {0: 'Y'})
```

```python
scaler = StandardScaler()
scaler = scaler.fit(train[['Y']])
train['Y'] = scaler.transform(train[['Y']])
test['Y'] = scaler.transform(test[['Y']])
```

```python
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
```

```python
velocity_data = pd.DataFrame(columns = ['Y', 'Z1', 'Z2'])
velocity_data['Y'] = np.mean(vel_Y[0][0], axis = 1)
velocity_data['Z1'] = np.mean(vel_Z1[0][0], axis = 1)
velocity_data['Z2'] = np.mean(vel_Z2[0][0], axis = 1)
velocity_data.index.name = 'Time'
```

```python
def central_diff(data):
    central_differences = []
    for i in range(1, len(data)-1):
        central_differences.append((data[i-1]-data[i+1])/2)
    return central_differences
```

```python
u = list(np.mean(vel_Y[0][0], axis = 1))
du = central_diff(u)
d2u = central_diff(du)
velocities = u[2:-2]
first_diff = du[1:-1]
fig = plt.figure(figsize=(12,8.4))
ax = fig.add_subplot(111, projection='3d');
ax.scatter(velocities, first_diff, d2u, c = 'black');
ax.set_xlabel('u');
ax.set_ylabel('du');
ax.set_zlabel('d2u');
```

```python
#calculating parameters
std_u = np.std(u)
std_du = np.std(du)
std_d2u = np.std(d2u)
u_max_u = std_u*(2*np.log(len(u)))**0.5
u_max_du = std_du*(2*np.log(len(du)))**0.5
u_max_d2u = std_d2u*(2*np.log(len(d2u)))**0.5
ang_of_rotation = np.arctan(sum(np.array(velocities)*np.array(d2u))/sum(np.array(velocities)**2))
```
